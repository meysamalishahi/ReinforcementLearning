# Computing the state-value of $4\times 4$ Gridworld, Example 4.1, page 76

Here, we compute the state-value function via direct approach, not using DP based on optimality equations.

```python
import numpy as np 
```

```python
M = np.zeros((15,15))
G = np.arange(16).reshape(4,4)
G[3,3] = 0
H = np.zeros((6,6))
H[1:5,1:5] = G
H[0] = H[1]
H[5] = H[4]
H[1:5, 0] = H[1:5, 1]
H[1:5, 5] = H[1:5, 4]
H = H.astype(int)
H
```

```python
for i in range(1,5):
    for j in range(1,5):

        M[H[i,j],H[i+1,j]] += 1 
        M[H[i,j],H[i-1,j]] += 1 
        M[H[i,j],H[i,j+1]] += 1
        M[H[i,j],H[i,j-1]] += 1
```

```python
M
```

```python
M[0]*=0 
M/=4
```

```python

M.sum(axis = 1)
```

```python
sum = np.zeros((15,15))
temp = M.copy()
for i in range(1,1000):
    sum += i * temp
    temp = temp @ M
```
